```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

spring_dataset = pd.read_csv('/Users/dami/Desktop/pythonTest/nailPJ/pccs_lab/spring.csv', index_col = 0)
summer_dataset = pd.read_csv('/Users/dami/Desktop/pythonTest/nailPJ/pccs_lab/summer.csv', index_col = 0)
fall_dataset = pd.read_csv('/Users/dami/Desktop/pythonTest/nailPJ/pccs_lab/fall.csv', index_col = 0)
winter_dataset = pd.read_csv('/Users/dami/Desktop/pythonTest/nailPJ/pccs_lab/winter.csv', index_col = 0)

spring_dataset.drop(spring_dataset.columns[spring_dataset.columns.str.contains('unnamed',case = False)],axis = 1, inplace = True)
summer_dataset.drop(summer_dataset.columns[summer_dataset.columns.str.contains('unnamed',case = False)],axis = 1, inplace = True)
fall_dataset.drop(fall_dataset.columns[fall_dataset.columns.str.contains('unnamed',case = False)],axis = 1, inplace = True)
winter_dataset.drop(winter_dataset.columns[winter_dataset.columns.str.contains('unnamed',case = False)],axis = 1, inplace = True)
spring_dataset.columns = ['l','a','b']
summer_dataset.columns = ['l','a','b']
fall_dataset.columns = ['l','a','b']
winter_dataset.columns = ['l','a','b']

print(spring_dataset.columns)
print(summer_dataset.columns)
print(fall_dataset.columns)
print(winter_dataset.columns)
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
spring_x = spring_dataset['l']
spring_y = spring_dataset['a']
spring_z = spring_dataset['b']

ax.scatter(spring_x, spring_y, spring_z)
ax.set_xlabel('L*')
ax.set_ylabel('a*')
ax.set_zlabel('b*')

plt.suptitle('scatter of spring dataset')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
summer_x = summer_dataset['l']
summer_y = summer_dataset['a']
summer_z = summer_dataset['b']

ax.scatter(summer_x, summer_y, summer_z)
ax.set_xlabel('L*')
ax.set_ylabel('a*')
ax.set_zlabel('b*')

plt.suptitle('scatter of summer dataset')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
fall_x = fall_dataset['l']
fall_y = fall_dataset['a']
fall_z = fall_dataset['b']

ax.scatter(fall_x, fall_y, fall_z)
ax.set_xlabel('L*')
ax.set_ylabel('a*')
ax.set_zlabel('b*')

plt.suptitle('scatter of fall dataset')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
winter_x = winter_dataset['l']
winter_y = winter_dataset['a']
winter_z = winter_dataset['b']

ax.scatter(winter_x, winter_y, winter_z)
ax.set_xlabel('L*')
ax.set_ylabel('a*')
ax.set_zlabel('b*')

plt.suptitle('scatter of winter dataset')
plt.show()
```

```python
spring_dataset['season'] = 'spring'
summer_dataset['season'] = 'summer'
fall_dataset['season'] = 'fall'
winter_dataset['season'] = 'winter'

combined_dataset = pd.concat([spring_dataset, summer_dataset, fall_dataset, winter_dataset])
print(combined_dataset)
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

colors = {'spring': 'hotpink', 'summer': 'steelblue', 'fall': 'darkorange', 'winter': 'darkgray'}

for season, data in combined_dataset.groupby('season'):
    ax.scatter(data['l'], data['a'], data['b'], c=colors[season], label=season)

ax.set_xlabel('L*')
ax.set_ylabel('a*')
ax.set_zlabel('b*')
plt.suptitle('Combined scatter of all seasons')
plt.legend()
plt.show()
```
