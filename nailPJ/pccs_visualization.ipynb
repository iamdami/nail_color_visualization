```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

spring_dataset = pd.read_csv('/Users/dami/Desktop/pythonTest/nailPJ/pccs_lab/spring.csv', index_col = 0)
summer_dataset = pd.read_csv('/Users/dami/Desktop/pythonTest/nailPJ/pccs_lab/summer.csv', index_col = 0)
fall_dataset = pd.read_csv('/Users/dami/Desktop/pythonTest/nailPJ/pccs_lab/fall.csv', index_col = 0)
winter_dataset = pd.read_csv('/Users/dami/Desktop/pythonTest/nailPJ/pccs_lab/winter.csv', index_col = 0)

spring_dataset.drop(spring_dataset.columns[spring_dataset.columns.str.contains('unnamed',case = False)],axis = 1, inplace = True)
summer_dataset.drop(summer_dataset.columns[summer_dataset.columns.str.contains('unnamed',case = False)],axis = 1, inplace = True)
fall_dataset.drop(fall_dataset.columns[fall_dataset.columns.str.contains('unnamed',case = False)],axis = 1, inplace = True)
winter_dataset.drop(winter_dataset.columns[winter_dataset.columns.str.contains('unnamed',case = False)],axis = 1, inplace = True)
spring_dataset.columns = ['l','a','b']
summer_dataset.columns = ['l','a','b']
fall_dataset.columns = ['l','a','b']
winter_dataset.columns = ['l','a','b']

print(spring_dataset.columns)
print(summer_dataset.columns)
print(fall_dataset.columns)
print(winter_dataset.columns)
```
