```python
import numpy as np
import pandas as pd

result_data = []

for i in range(1, 101):
    data_name = f"m{i}"
    
    if data_name in bandi_dataset.index:
        nail_lab = bandi_dataset.loc[data_name, ['L', 'a', 'b']].values
    else:
        print(f"Invalid data name: {data_name}")
        continue
    
    print(f"{data_name}, [L*, a*, b*]: {nail_lab}")
    
    visited_indices = set()
    closest_b_difference = np.inf
    closest_season_data = None
    
    similar_color_found = False  # 유사색이 발견되었는지 여부를 나타내는 플래그 추가
    
    for season, data in combined_dataset.groupby('season'):
        season_lab = data[['l', 'a', 'b']].values

        # 계절별로 b값 조정해 거리 계산
        if season in ['summer', 'winter']:
            season_lab[:, 2] = season_lab[:, 2] + 5  # summer과 winter인 경우 b 값을 -5로 조정
        elif season in ['fall', 'spring']:
            season_lab[:, 2] = season_lab[:, 2] - 5  # fall인 경우 b 값을 +5로 조정
        
        b_difference = np.abs(season_lab[:, 2] - nail_lab[2])  # 반디 네일 데이터의 b 값과 각 계절 데이터의 b 값 차이 계산
        
        min_b_difference_indices = np.argsort(b_difference)[:2]  # 가장 작은 두 개의 b 값 차이를 가지는 데이터의 인덱스
        
        closest_b_values = season_lab[min_b_difference_indices, 2]  # 가장 가까운 두 개의 b 값
        
        if np.abs(nail_lab[2] - closest_b_values[0]) < np.abs(nail_lab[2] - closest_b_values[1]):  # 첫 번째 값이 더 가까운 경우
            closest_b_value = closest_b_values[0]
            closest_b_index = min_b_difference_indices[0]
        else:  # 두 번째 값이 더 가까운 경우
            closest_b_value = closest_b_values[1]
            closest_b_index = min_b_difference_indices[1]
            
        if np.abs(nail_lab[2] - closest_b_value) < closest_b_difference:  # 현재 계절과 반디 네일 데이터 간의 b 값 차이가 더 작을 때
            closest_b_difference = np.abs(nail_lab[2] - closest_b_value)  # 가장 작은 b 값 차이 갱신
            closest_season_data = data.iloc[closest_b_index]  # 가장 작은 b 값 차이를 가지는 계절 데이터 갱신
                
    similar_color_found = False  # 유사색이 발견되었는지 여부 초기화
    
    visited_indices = set()  # visited_indices 초기화
    
    # 동일색 및 유사색을 찾기 위한 루프
    for season, data in combined_dataset.groupby('season'):
        season_lab = data[['l', 'a', 'b']].values
        
        # 계절별로 b값 조정
        if season in ['summer', 'winter']:
            season_lab[:, 2] = season_lab[:, 2] + 5
        elif season in ['fall', 'spring']:
            season_lab[:, 2] = season_lab[:, 2] - 5
        
        distance = np.abs(season_lab - nail_lab)
        
        for idx, row in enumerate(np.abs(distance)):
            if idx in visited_indices:
                continue
            
            if np.abs(distance[idx][0]) <= 3 and np.abs(distance[idx][1]) <= 3 and np.abs(distance[idx][2]) <= 3:
                label = "Same Color"
                if not similar_color_found:  # 유사색이 발견되지 않은 경우에만 출력
                    print(f"{label}: {data.index[idx]}, L*, a*, b*: {season_lab[idx]}, Season: {season}")
                    result_data.append([data_name, 'Same Color', data.index[idx], season, *season_lab[idx]])
                visited_indices.add(idx)
                if len(visited_indices) == 3:
                    break
            elif np.abs(distance[idx][0]) <= 5 and np.abs(distance[idx][1]) <= 5 and np.abs(distance[idx][2]) <= 5:
                label = "Similar Color"
                if not similar_color_found:  # 유사색이 발견되지 않은 경우에만 출력
                    print(f"{label}: {data.index[idx]}, L*, a*, b*: {season_lab[idx]}, Season: {season}")
                    result_data.append([data_name, 'Similar Color', data.index[idx], season, *season_lab[idx]])
                    similar_color_found = True
                visited_indices.add(idx)
                if len(visited_indices) == 3:
                    break
                    
    if closest_season_data is not None:
        if not similar_color_found:  # 동일색 또는 유사색이 발견되지 않은 경우에만 출력
            print(f"PCCS Color: {closest_season_data.name}, [L*, a*, b*]: {closest_season_data[['l', 'a', 'b']].values}, The Closest Season: {closest_season_data['season']}")
            result_data.append([data_name, 'Closest Season', closest_season_data.name, closest_season_data['season'], *nail_lab])
    
    print()
    print()

result_df = pd.DataFrame(result_data, columns=['Data Name', 'Type', 'Color Name', 'Season', 'L*', 'a*', 'b*'])
result_df.to_csv('/Users/dami/Desktop/pythonTest/nailPJ/pccs_lab/res1.csv', index=False)
```

```python
from colormath.color_objects import LabColor, sRGBColor
from colormath.color_conversions import convert_color
import pandas as pd
import matplotlib.pyplot as plt

# HVC 데이터셋을 불러옵니다.
hvc_lab_data = pd.read_csv('/Users/dami/Desktop/pythonTest/nailPJ/hvc_lab_data.csv', index_col=0)

# 각 데이터 포인트에 대해 RGB 색상값을 계산합니다.
def lab_to_rgb(L, a, b):
    lab = LabColor(L, a, b)
    rgb = convert_color(lab, sRGBColor)
    # RGB 값을 0에서 1 사이의 범위로 스케일링합니다.
    rgb_r = max(min(rgb.rgb_r, 1), 0)
    rgb_g = max(min(rgb.rgb_g, 1), 0)
    rgb_b = max(min(rgb.rgb_b, 1), 0)
    return [rgb_r, rgb_g, rgb_b]


# 각 데이터 포인트의 RGB 색상값을 계산합니다.
rgb_colors = [lab_to_rgb(row['L'], row['a'], row['b']) for _, row in hvc_lab_data.iterrows()]

# RGB 색상값을 시각화하여 데이터셋을 산점도로 표현합니다.
plt.figure(figsize=(20, 1))
for i, rgb_color in enumerate(rgb_colors):
    plt.scatter(i, 1, c=[rgb_color], s=3000, marker='s')

plt.title('HVC Colors')
plt.axis('off')
plt.show()
```
